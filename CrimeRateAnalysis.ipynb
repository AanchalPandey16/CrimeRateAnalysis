#Question 1: What is the total number of rape cases in a specific year?

```python
plt.figure(figsize=(7, 4))
rape_cases_trend = df.groupby('Year')['Rape Cases'].sum()
sns.lineplot(x=rape_cases_trend.index, y=rape_cases_trend.values, marker='o', color='Purple', linewidth=2)
plt.title("Total Rape Cases per Year")
plt.xlabel("Year")
plt.ylabel("Total Rape Cases")
plt.show()
```

# Question 2: Which state has the highest crime rate in a given year?


```python
plt.figure(figsize=(7, 4))
top_states_by_crime_rate = df.sort_values('Crime Rate', ascending=False).head(10)
sns.barplot(x='Crime Rate', y='State', data=top_states_by_crime_rate, hue = 'State',palette='Purples')
plt.title("Top States by Crime Rate")
plt.xlabel("Crime Rate")
plt.ylabel("State")
plt.show()
```

# Question 3: What is the correlation between total crimes and population?


```python
plt.figure(figsize=(8, 4))
sns.scatterplot(x='Population', y='Total Crimes', data=df, hue='State', size='Total Crimes', sizes=(20, 200), palette="Purples", legend=False)
plt.title("Total Crimes vs. Population")
plt.xlabel("Population")
plt.ylabel("Total Crimes")
plt.show()
```

# Question 4: What is the average crime clearance rate vary across different states?

```python
plt.figure(figsize=(15, 5))
state_avg_clearance_rate = df.groupby('State')['Crime Clearance Rate'].mean().sort_values(ascending=False)
sns.barplot(x=state_avg_clearance_rate.index, y=state_avg_clearance_rate.values, hue=state_avg_clearance_rate.index, palette='Purples', dodge=False)
plt.title("Average Crime Clearance Rate by State")
plt.xlabel("State")
plt.ylabel("Average Crime Clearance Rate")
plt.xticks(rotation=90)
plt.show()
```

# Question 5: What is the distribution of abduction cases across districts?

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Abduction Cases', y='District', data=df, color='#D8B5D7')
plt.title("Distribution of Abduction Cases by District")
plt.xlabel("Abduction Cases")
plt.ylabel("District")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Question 6: What is the trend in the number of murder cases over the years?


```python
murder_trend = df.groupby('Year')['Murder Cases'].sum()
plt.figure(figsize=(8, 5))
sns.lineplot(x=murder_trend.index, y=murder_trend.values, marker='o', color='#D8B5D7', linestyle='--', linewidth=2)
plt.title("Trend in Murder Cases Over Years")
plt.xlabel('Year')
plt.ylabel('Murder Cases')
plt.show()
```

# Question 7: Which district has the highest number of arrests in a given year?


```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Arrests', y='District', data=df, hue='District', palette='Purples')
plt.title("Districts with Highest Number of Arrests")
plt.xlabel("Number of Arrests")
plt.ylabel("District")
plt.show()
```

# Question 8: What is the average crime rate across all states for a particular year?


```python
plt.figure(figsize=(10, 6))
avg_crime_rate_by_state = df.groupby(['Year', 'State'])['Crime Rate'].mean().reset_index()
sns.lineplot(x='Year', y='Crime Rate', hue='State', data=avg_crime_rate_by_state, palette='Purples', marker='o')
plt.title("Average Crime Rate Across States Over the Years")
plt.xlabel("Year")
plt.ylabel("Average Crime Rate")
plt.legend(title='State')
plt.xticks(rotation=45)
plt.show()
```

# Question 9:  Calculate the total number of cases for each crime type



```python
crime_type_counts = df[['Rape Cases', 'Abduction Cases', 'Murder Cases']].sum()
crime_type_counts.plot(kind='bar', stacked=True, color=['#b5a0d0', '#9b8cc4', '#6f4c9e'], figsize=(8, 6))
plt.title("Crime Type Distribution")
plt.xlabel("Crime Types")
plt.ylabel("Number of Cases")
plt.xticks(rotation=0)
plt.show()
```

# Question 10: Correlation Matrix of Crime Data



```python
correlation_matrix = df[['Rape Cases', 'Abduction Cases', 'Murder Cases', 'Total Crimes', 'Population', 'Crime Rate', 'Arrests']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='Purples', fmt='.2f', linewidths=0.5, vmin=-1, vmax=1)
plt.title("Correlation Matrix of Crime Data")
plt.show()
```
