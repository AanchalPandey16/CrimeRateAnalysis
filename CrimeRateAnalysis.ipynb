```python
# Find the total number of Rapes cases over the years
plt.figure(figsize=(7, 4))
# Group the dataset by 'Year' and calculate the sum of 'Rape Cases' for each year
rape_cases_trend = df.groupby('Year')['Rape Cases'].sum()

# Plot the line chart showing the trend of total rape cases per year
sns.lineplot(x=rape_cases_trend.index, y=rape_cases_trend.values, marker='o', color='Purple', linewidth=2)

# Title and labels for the plot
plt.title("Total Rape Cases per Year")
plt.xlabel("Year")
plt.ylabel("Total Rape Cases")

# Show the plot
plt.show()

print("Conclusion")
print("The line graph shows the total number of rape cases reported over a period from 2018 to 2022.\nShows sharp incline in cases from 2018-19 and then significant decrease in reported cases from 2019-20 ")
```

```python
# Find out which state has highest crime rate
plt.figure(figsize=(7, 4))

# Sort the dataset by 'Crime Rate' in descending order and get the top 10 states
top_states_by_crime_rate = df.sort_values('Crime Rate', ascending=False).head(10)

# Create a bar plot to visualize the top 10 states by crime rate
sns.barplot(x='Crime Rate', y='State', data=top_states_by_crime_rate, color='#C3B1E1')

# Set title and labels for the plot
plt.title("Top States by Crime Rate")
plt.xlabel("Crime Rate")  # Label for the x-axis (Crime Rate)
plt.ylabel("State")  # Label for the y-axis (States)

# Display the plot
plt.show()

print("Conclusion")
print("The bar chart represents the crime rates across top five states. \nAs in the visual Kerala has the highest crime rate among all the other state. ")
```

```python
plt.figure(figsize=(8, 4))
# Create a scatter plot with Population on the x-axis and Total Crimes on the y-axis
sns.scatterplot(x='Population', y='Total Crimes', data=df, hue='State', size='Total Crimes', sizes=(20, 200), palette="Purples", legend=False)

# Set title and labels for the plot
plt.title("Total Crimes vs. Population")
plt.xlabel("Population")
plt.ylabel("Total Crimes")

# Show the plot
plt.show()

print("Conclusion")
print("The graph shows the correlation between total crimes and population. Each dot represents an area.\n The dots are scattered, showing no clear relationship between population size and the number of crimes.\n This means that population alone doesn't fully explain how many crimes happen in a place.")
```

```python
# Find out the average crime clearance rate vary across different states
plt.figure(figsize=(10, 5))
# Group the data by 'State' and calculate the mean 'Crime Clearance Rate' for each state, then sort them in descending order
state_avg_clearance_rate = df.groupby('State')['Crime Clearance Rate'].mean().sort_values(ascending=False).head(20)

# Create a bar plot using the calculated average crime clearance rates
sns.barplot(x=state_avg_clearance_rate.index, y=state_avg_clearance_rate.values, color='#C3B1E1', dodge=False)

# Set the title and labels for the plot
plt.title("Average Crime Clearance Rate by State")
plt.xlabel("State")
plt.ylabel("Average Crime Clearance Rate")

# Rotate x-axis labels (state names) to avoid overlap and ensure readability0
plt.xticks(rotation=90)

# Show the plot
plt.show()

print("Conclusion")
print("This bar chart shows how well different states in India solve crimes.\nUttar Pradesh has the highest clearance rate where as Delhi has the lowest.")
```

```python
# Analyse the distribution of abduction cases in western states
# Defining the western states
western_states = ['Goa', 'Gujarat', 'Maharashtra', 'Rajasthan']

# Filtering the dataset for western states
western_states_df = df[df['State'].isin(western_states)]

# Plot the boxplot for Abduction Cases in Western States
plt.figure(figsize=(12, 6))
sns.boxplot(x='State', y='Abduction Cases', data=western_states_df, color='#C3B1E1')

plt.title("Distribution of Abduction Cases in Western States")
plt.xlabel("State")
plt.ylabel("Abduction Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("Conclusion")
print("This box plot shows abduction cases across four different states in India.\nThe length of the boxes varies, indicating different ranges of cases. Some show a positive skew, suggesting more State with lower numbers of cases.\nGujarat shows the highest median abduction cases, while Goa has the lowest.\nThe distributions for all states are relatively symmetrical, suggesting similar patterns in the spread of abduction cases.")
```

```python
# Analyse the trend in the number of murder cases over the years
murder_trend = df.groupby('Year')['Murder Cases'].sum()
plt.figure(figsize=(8, 5))

# Create a line plot with murder cases trend
sns.lineplot(x=murder_trend.index, y=murder_trend.values, marker='o', color='#C3B1E1', linestyle='--', linewidth=2)

# Set the title and labels for the plot
plt.title("Trend in Murder Cases Over Years")
plt.xlabel('Year')
plt.ylabel('Murder Cases')

# Show the plot
plt.show()

print("Conclusion")
print("The data shows a slight increase in murder cases from 2018 to 2019, followed by a significant spike in 2020.\nHowever, the number of murder cases then dropped sharply in 2021 and continued to decrease slightly in 2022.\n This graph suggests that while there was a concerning peak in murder cases in 2020, there has been a subsequent decline in recent years.")
```

```python
# Find out which state has highest number of arrest
plt.figure(figsize=(10, 6))

#pick top 10 district
top_districts = df.sort_values(by='Arrests', ascending=False).head(10)

#barplot for the top 10 districts
sns.barplot(x='Arrests', y='State', data=top_districts, color='#C3B1E1')

# Set the title and labels
plt.title("Top 10 Districts with Highest Number of Arrests")
plt.xlabel("Number of Arrests")
plt.ylabel("District")
plt.show()

print("Conclusion")
print("The bar plot represents the number of arrests in different districts in India.\nUttar Pradesh has the  highest number of arrests, while Arunachal Pradesh has the lowest.")
```

```python
# Analyse the average crime rate across southern states over the years
#Southern States
southern_states = ['Andhra Pradesh', 'Telangana', 'Tamil Nadu', 'Karnataka', 'Kerala', 'Puducherry']

# Filter the dataset for Southern States
southern_states_df = df[df['State'].isin(southern_states)]

# Calculate the average crime rate by state and year
avg_crime_rate_by_state = southern_states_df.groupby(['Year', 'State'])['Crime Rate'].mean().reset_index()

# Plot the line plot for Southern States
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Crime Rate', hue='State', data=avg_crime_rate_by_state, palette='cool', marker='o')

# Set title and labels
plt.title("Average Crime Rate Across Southern States Over the Years")
plt.xlabel("Year")
plt.ylabel("Average Crime Rate")
plt.legend(title='State')
plt.xticks(rotation=45)

# Show the plot
plt.show()

print("Conclusion")
print("The graph shows how the average crime rate changed in  southern Indian states from 2018 to 2022.\nSome states like Andhra Pradesh and Telangana saw a big rise in crime at first, then it went down.\nOther states like Karnataka and Tamil Nadu had smaller changes in their crime rates.\nPuducherry and Kerala had more ups and downs, with the crime rate going up and down a lot.")
```

```python
# Total number of cases for each crime type
# Sum the cases for each crime type
crime_type_counts = df[['Rape Cases', 'Abduction Cases', 'Murder Cases']].sum()

# Plot the donut chart
plt.figure(figsize=(8, 6))
plt.pie(crime_type_counts, labels=crime_type_counts.index, autopct='%1.1f%%', colors=['#9b8cc4', '#6f4c9e', '#b5a0d0'], wedgeprops=dict(width=0.4))

# Set the title
plt.title("Crime Type Distribution")

# Show the plot
plt.show()

print("Conclusion")
print("This donut chart shows the number of cases reported for different types of crimes.\nAbduction cases are the most frequently reported crime type, followed by rape cases, and then murder cases.\nThere's a significant difference in the number of cases reported for each crime type.")
```

```python
# Correlation Matrix of Crime Data
# Ensure the relevant columns are numeric
numeric_columns = df[['Rape Cases', 'Abduction Cases', 'Murder Cases', 'Total Crimes', 'Population', 'Crime Rate', 'Arrests']]

# Convert columns to numeric, forcing errors to NaN
numeric_columns = numeric_columns.apply(pd.to_numeric, errors='coerce')

# Calculate the correlation matrix
correlation_matrix = numeric_columns.corr()

# Set up the visual style of the plot
plt.figure(figsize=(10, 5))  # Adjust the size to make the plot clearer
sns.set(style='white')  # Use a clean white background

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='Purples', fmt='.2f', linewidths=0.8, vmin=-1, vmax=1,
            cbar_kws={'label': 'Correlation Coefficient'}, square=True)

# Set the title
plt.title("Correlation Matrix of Crime Data", fontsize=16)

# Display the plot
plt.show()

print("Conclusion")
print("This visual shows the relationships between different types of crimes, population, crime rate, and arrests.\nThe darker colors in the boxes show stronger correlation and light colors it shows weaker correlation.")
```
